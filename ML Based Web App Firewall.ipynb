```python
import pandas as pd
http_request_df = pd.read_csv("processed_urls.csv")
```

```python
from pycaret.classification import *
from sklearn.model_selection import train_test_split
```

```python
train_set, test_set = train_test_split(http_request_df.loc[:8000,:])
```

```python
clu1 = setup(data = train_set, target='type', \
             numeric_features=['unicode', 'inequality', 'pipe', 'semicolon', 'curlybrace', 'caret', 'consecutive_num', 'hyphen', 'underscore'])
```

```python
best_model = compare_models()
```

```python
evaluate_model(best_model)
```

```python
results = predict_model(best_model, test_set)
```

```python
test_set
```

```python
correct_pred_count = 0
for index in range(len(results['Label'])):
    if results['Label'].iloc[index] == results['type'].iloc[index]:
        correct_pred_count += 1

print(correct_pred_count / len(results['Label'])) # About 79.56% prediction accuracy using a random forest classifer
```
